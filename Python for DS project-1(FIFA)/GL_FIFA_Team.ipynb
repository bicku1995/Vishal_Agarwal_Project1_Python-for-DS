# 4. Convert the columns "Value", "Wage", "Release Clause" to float datatype after getting rid of currency symbol and suffix. 

- Note: When the record/entry has "M"(indicates millions) as suffix you
need to multiply that value with 1000000
- When the record/entry has "K"(indicates thousands) as suffix you
need to multiply that value with 1000


```python
df=data.copy()
#Creating a copy of dataset
```

```python
df[['Value','Wage','Release Clause']]=df[['Value','Wage','Release Clause']].replace({'€':''},regex=True)
#removing pound sign 
```

```python
def curr_mul(x):
    if type(x)==float or type(x)==int:
        return x
    if 'K' in x:
        return float(x.replace('K',''))*1000
    if 'M' in x:
        return float(x.replace('M',''))*1000000
    return 0.0        
 #creating function to remove sufix like M ,K and converting to float datatype
```

```python
df['Release Clause']=df['Release Clause'].fillna(0)
df['Release Clause']=df['Release Clause'].apply(curr_mul)
df['Wage']=df['Wage'].apply(curr_mul)
df['Value']=df['Value'].apply(curr_mul)
```

```python
df.head().T
```

# 7. The column 'Height' is in inches with a quotation mark, Convert to float with decimal points.

```python
df['Height'].head()
```

```python
def height_convert(x):
    if(type(x)==str):
        f=int(x.split("'")[0])
        i=int(x.split("'")[1])               
        t=float((f*12)+i)
        return t
```

```python
df['Height']=df['Height'].apply(height_convert)
```

```python
df['Height'].head()
```

# 8. The column "Weight" has the suffix as lbs, remove the suffix and convert to float. 

```python
df['Weight'].head()
```

```python
def lbs(x):
    if(type(x)==str):
        if 'lbs' in x:
            return float(x.replace('lbs',''))
```

```python
df['Weight']=df['Weight'].apply(lbs)
df['Weight'].head()
```
